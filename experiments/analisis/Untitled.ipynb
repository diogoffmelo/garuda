```python
import pickle
import os
import numpy as np
import pandas as np
import matplotlib.pyplot as plt

```

```python
pickle_files = ['./' + n for n in  os.listdir('./') if '.pkl' in n]
```

```python
def open_report(pickle_file, vesbose=True):
    with open(pickle_file, 'br') as f:
        report = pickle.load(f)

    if vesbose:
        print('\nfile_name: {}\nmodel:{}'.format(pickle_file, report['model']))
    
    return report
```

```python
for fname in pickle_files:
    open_report(fname)
```
